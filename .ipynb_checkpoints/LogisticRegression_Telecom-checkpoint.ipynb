```python
def handleDummies(dftemp,col,dropCol,prefix):
    print(dftemp.shape)
    dum=pd.get_dummies(dftemp[col],prefix)
    dropCol=col+'_'+dropCol
    dum=dum.drop(dropCol,axis=1)
    dftemp=pd.concat([dftemp,dum],axis=1)
    dftemp=dftemp.drop(col,axis=1)
    return dftemp
```

```python
catcols=[]
catDict={}
valCheck='No internet service'
for col in df.columns:
    if len(df[col].unique())<5 and df[col].dtype=='O':
        print(col,':',df[col].unique())
        catcols.append(col)
        if 'No internet service' in df[col].unique():
            catDict[col]='No internet service'
        elif 'Credit card (automatic)' in df[col].unique():
            catDict[col]='Credit card (automatic)'
        elif 'Two year' in df[col].unique():
            catDict[col]='Two year'
        elif 'No phone service' in df[col].unique():
            catDict[col]='No phone service'
        elif 'Male' in df[col].unique():
            catDict[col]='Male'
        elif 'No' in df[col].unique():
            catDict[col]='No'
        else:
            print('end')

catDict            
```

```python
for col,val in catDict.items():
    df=handleDummies(df,col,val,col)
    print(col,val)
```

```python
num=[float(x)/10 for x in range(10)]
for i in num:
    y_train_pred_final[i]=y_train_pred_final.Churn_Prob.map(lambda x: 1 if x>i else 0)
y_train_pred_final.head()
```

```python
cutoof_df=pd.DataFrame(columns=['prob','accuracy','sensi','speci'])
num
for i in num:
    cml=metrics.confusion_matrix(y_train_pred_final.Churn,y_train_pred_final[i])
    total=sum(sum(cml))
    accuracy=((cml[0,0]+cml[1,1])/total)
    sensi=cml[0,0]/(cml[0,0]+cml[0,1])
    speci=cml[1,1]/(cml[1,0]+cml[1,1])
    cutoof_df.loc[i]=[i,accuracy,sensi,speci]
print(cutoof_df)
```

```python
cutoof_df.plot.line(x='prob',y=['accuracy','sensi','speci'])
```
